```python
data=data.values
```

```python
dist_mat = np.zeros((5,299))
```

```python
def eclid_distance(centroid):
    for i in range(len(centroid)):
        for j in range(data.shape[0]):
          dist_mat[i][j] =np.sqrt(np.sum(np.square(centroid[i]-data[j:j+1,])))       
    return(dist_mat)
```

```python
def kmean(k):
    c=np.random.randint(0,data.shape[0],size=k)
    centroid=[]
    for i in range(len(c)):
        centroid.append(data[c[i]])
    dist_mat=eclid_distance(centroid)
    for i in range(10):
        x = [[] for _ in range(k)]
        for i in range(data.shape[0]):
          for j in range(k):
            if (dist_mat[:,i])[j]==dist_mat[:,i].min():
                x[j].append(data[i])
        for i in range(k):
            x[i]=np.array(x[i])
            centroid[i]=x[i].mean(axis=0)
    return(x) 
```
